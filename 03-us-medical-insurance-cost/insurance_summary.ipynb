```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
# Data Import
df_main = pd.read_csv('insurance_cleaned.csv')
df_main.info()
```

```python
# Data Preview
df_main.head()

# Generate BMI Categories
def generate_bmi_categories(bmi):
    if bmi < 19.5:
        return "Underweight"
    elif 19.5 < bmi <= 24.5:
        return "Normal"
    elif 24.5 < bmi <= 29.5:
        return "Overweight"
    elif 29.5 < bmi <= 39.2:
        return "Obese"
    elif 39.2 < bmi <= 52.3:
        return "Morbidly Obese"
    else:
        return np.nan

df_main['BMI Category'] = df_main['bmi'].apply(generate_bmi_categories)
```

```python
df_main.head(25)
```

```python
##### Descriptive Analysis
# Age Dist
observations = df_main.age.count()
average_age = df_main.age.mean()
median_age = df_main.age.median()
age_dist_plot = sns.histplot(df_main.age, bins=8, color='orange')
plt.axvline(average_age, label='Average', color='darkorange', linestyle='--')
plt.axvline(median_age, label='Median', color='yellow', linestyle='--')
plt.title('Age Distribution')
plt.legend()
plt.show()
plt.clf()

# Sex Dist 
sex_prop = df_main.sex.value_counts(normalize=True)
plt.pie(df_main.sex.value_counts(), autopct='%1.1f%%', labels=['Male', 'Female'])
plt.title("Sex Distribution")
plt.legend()
plt.show()

# BMI Dist
average_bmi = df_main.bmi.mean()
median_bmi = df_main.bmi.median()
bmi_dist_plot = sns.histplot(df_main.bmi, bins=8, color='green')
plt.axvline(average_bmi, label='Average', color='darkgreen', linestyle='--')
plt.axvline(median_bmi, label='Median', color='lightgreen', linestyle='--')
plt.title('BMI Distribution')
plt.legend()
plt.show()
plt.clf()

# Children Dist
average_chil = df_main.children.mean()
median_chil = df_main.children.median()
children_dist_plot = sns.histplot(df_main.children, bins=5, color='beige')
plt.axvline(average_chil, label='Average', color='brown', linestyle='--')
plt.axvline(median_chil, label='Median', color='tan', linestyle='--')
plt.title('Children Distribution')
plt.legend()
plt.show()
plt.clf()

# Region Dist
region_dist_plot = plt.pie(df_main.region.value_counts(), labels=set(df_main.region), autopct='%1.1f%%')
plt.title('Region Distribution')
plt.show()
plt.clf()

# Charges Dist
average_cost = df_main.charges.mean()
median_cost = df_main.charges.median()
charges_plt = sns.histplot(df_main.charges, bins=20, color='green')
plt.axvline(average_cost, label='Average', color='darkgreen', linestyle='--')
plt.axvline(median_cost, label='Median', color='limegreen', linestyle='--')
plt.title('Charge Distribution')
plt.legend()
plt.show()
plt.clf()
```

```python
# Exploratory
print(df_main.head())

sns.lmplot(data=df_main, x='smoker', y='charges', hue='sex')
plt.title("Difference in Charges between Smoker and Non-Smokers")
plt.show()
plt.clf()

sns.pointplot(data=df_main, x='region', y='bmi', hue='sex')
plt.title("BMI of Different US Regions")
plt.show()
plt.clf()

sns.pointplot(data=df_main, x='region', y='charges', hue='sex')
plt.title("Insurance Prices of US Regions")
plt.show()
plt.clf()

sns.lmplot(data=df_main, x='age', y='charges', hue='BMI Category', palette={'Underweight':"#0679d1", 'Normal':"#f9ebe5", 'Overweight':"#cc6452", 'Obese':"#c83113", 'Morbidly Obese' :"#ff2c00"})
plt.title("Insurance Prices by BMI Class")
plt.show()
plt.clf()

sns.pointplot(data=df_main, x='children', y='charges', hue='sex')
plt.title("Insurance Prices with Kids")
plt.show()
plt.clf()
```

# Conclusions

* Insurance distribution is equal across US regions and sex,
* Must individuals have about $10,000 in charges
* Smoking increases your premium
* Prices increase in the south, but may be attributed to the higher BMI's in the south
* Men typically pay higher premiums than women
* Number of children does not seem to affect insurance premiums
* BMI increases premiums
